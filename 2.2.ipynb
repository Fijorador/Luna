```python
frase=""


def ouvir_micro():
    time_sleep=0.1
    from keyword import iskeyword
    global frase
    microfone = sr.Recognizer()
    with sr.Microphone() as source:
        microfone.adjust_for_ambient_noise(source)
        sleep(time_sleep)
        print('Luna:'+'Estou ouvindo...')        
        audio=microfone.listen(source)
        sleep(time_sleep)
        try:            
            frase = str( microfone.recognize_google(audio,language='pt-BR'))
            if iskeyword(frase) is False:
                print('Humano:'+ frase)
                frase =str(microfone.recognize_google(audio,language='pt-BR'))           
        except sr.UnknownValueError:
                frase=("Som não captado")
                print('Humano:'+ frase)
        except:
                frase=("Erro não mapeado")
                print('Humano:'+ frase)
        
    return frase
```

```python
from pygame import mixer
import random
import os
from time import sleep

def cria_audio(som):
    time_sleep = 0.1
    r = random.randint(1,1000)
    op=1
    
    #reconhecimento 
    tts = gTTS(som, lang='pt-BR')
    print('traduzido')
    
    #salvar
    sleep( time_sleep )
    tts.save('audiotemp' + str(r) + '.mp3')
    print('salvo')
    
    #tocar
    while op!=0:
        try:
            playsound('audiotemp' + str(r) + '.mp3')
            op=0
        except:
            print("Aguarde(Lendo resposta Playsound).")
            op=1
        
    
    #limpar
    sleep( time_sleep )
    os.remove('audiotemp' + str(r) + '.mp3')
   
 
```
